```python
import pandas as pd
from dotenv import load_dotenv
from winterdrb.ztf import download_kowalski_alert_data
from tqdm import tqdm
from astropy.time import Time
```

```python
load_dotenv()
```

```python
df = pd.read_csv("~/new_real.csv")
n_tot = len(df)
```

```python
year = [int(x[3:5]) for x in df["ztf_xmatch"]]
df["ztfyear"] = [int(x[3:5]) for x in df["ztf_xmatch"]]
```

```python
# Get detection jd
row = df.iloc[0]
date = row["file"].split("/")[5]
dates = [x.split("/")[5] for x in df["file"]]
times = [Time(f"{date[:4]}-{date[4:6]}-{date[6:]}T00:00:00", format="isot") for date in dates]
jds = [x.jd for x in times]
df["winterjd"] = jds
df.sort_values(by="winterjd", ascending=False, inplace=True)
```

```python
# deduplicate, take only latest detection of each WNTR source
new = []
for name in tqdm(set(df["WINTER_name"])):
    matches = df[df["WINTER_name"] == name]
    new.append(matches.iloc[0])
df = pd.DataFrame(new)
```

```python
mask = (df["ztfyear"] > 22)
df = df[mask]
df
```

```python
mask = df["Type"].isin(['hostless', 'nuclear', 'offset from a galaxy'])
df = df[mask]
df
```

```python
# df = df[:200]
```

```python
df.reset_index(drop=True, inplace=True)

matches = []

for _, row in tqdm(df.iterrows(), total=len(df)):
    new = download_kowalski_alert_data(row["ztf_xmatch"])[0]
    matches.append(new["candidate"])
```

```python
match_df = pd.DataFrame(matches)
match_df
```

```python
df = pd.concat([df, match_df], axis=1)
df
```

```python
# ssdistnr
mask = (
    (df["isdiffpos"].isin(["1", "t", "true", "True", "T"])) # positive dets please
    & (df["drb"] > 0.5)  # drb good
    & (df["ndethist"] > 5)  # No asteroids
    & ((df["jdendhist"] - df["jdstarthist"]) > 0.01) # No asteroids, mk II
    & (df["winterjd"] > (df["jdstarthist"] -100.)) & ((df["jdendhist"] + 100.) >  df["winterjd"]) # winter dets lie near/between ZTF ones
)
df[mask]
```

```python
df.sort_values(by="ndethist", ascending=False, inplace=True)
df[mask][["WINTER_name", "ztf_xmatch", "ndethist"]]
```

```python
print(f"Fraction of confident reals {mask.sum()/n_tot}, have {mask.sum()} new reals")
```
